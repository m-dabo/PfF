```python
## Starting example
import numpy as np
np.random.seed(1000)
y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x, y)  ## or equivalently plt.plot(y)
```

```python
## 2-dimensional data sets
#np.random.seed(1000)
y = np.random.standard_normal((20, 2)).cumsum(axis = 0) ## calculate the cumulative sum of the sample data along the 1st axis
plt.figure(figsize = (6, 4))
plt.plot(y, lw = 1.5)
plt.plot(y, 'k*')
plt.title('happy stars')
```

```python
## Adding legend
plt.figure(figsize = (6, 4))
plt.plot(y[:, 0], lw = 1.5, label = 'first')
plt.plot(y[:, 1], lw = 1.5, label = 'second')
plt.plot(y, 'k*')
plt.legend(loc = 0) ## place a legend in the best location
plt.title('happy stars')
```

```python
## Resolving scaling issues by creating a new y-axis
y[:, 0] = y[:, 0] * 100 ## rescaling the data
fig, ax1 = plt.subplots() ## define the figure and axis object
plt.plot(y[:, 0], 'b', lw = 1.5, label = 'first')
plt.plot(y[:, 0], 'k*')
plt.ylabel('data1')
plt.legend(loc = 8) ## legend for 1st subplot in lower center
plt.grid(False)
ax2 = ax1.twinx() ## 2nd y-axis that shares the x-axis
plt.plot(y[:, 1], 'r', lw = 1.5, label = 'second')
plt.plot(y[:, 1], 'yh')
plt.ylabel('data2')
plt.legend(loc = 0) ## place a legend in the best location
plt.title('happy stars')
plt.grid(False)
```

```python
## Resolving scaling issues by creating a different subplot
plt.figure(figsize = (6, 4))
plt.subplot(211) ## define upper subplot with correponding coor.
plt.plot(y[:, 0], 'b', lw = 1.5, label = 'first')
plt.plot(y[:, 0], 'k*')
plt.ylabel('data1')
plt.legend(loc = 0) ## legend for 1st subplot in 'best' posi.
plt.grid(False)
plt.title('happy stars')
plt.subplot(2, 1, 2) ## lower subplot
plt.plot(y[:, 1], 'r', lw = 1.5, label = 'second')
plt.plot(y[:, 1], 'yh')
plt.ylabel('data2')
plt.legend(loc = 0) ## place a legend in the best location
plt.grid(False)
```

```python
## Combining a line plot with a bar chart
plt.figure(figsize = (6, 4))
plt.subplot(121) ## define upper subplot with correponding coor.
plt.plot(y[:, 0], 'b', lw = 1.5, label = 'first')
plt.plot(y[:, 0], 'k*')
plt.ylabel('value')
plt.xlabel('index')
plt.legend(loc = 0) ## legend for 1st subplot in lower center
plt.grid(False)
plt.title('happy stars')
plt.subplot(1, 2, 2) ## lower subplot
plt.bar(np.arange(len(y[:, 1])), y[:, 1], color = 'g', width = .5, label = 'second')
plt.xlabel('index')
plt.legend(loc = 0) ## place a legend in the best location
plt.grid(False)
plt.title('boring bars')
```

```python
## Scatter plot
y = np.random.standard_normal((1000, 2)) 
plt.figure(figsize = (6, 4))
plt.plot(y[:, 1], y [:, 0], 'go') ## same as plt.scatter(x, y, ...)
plt.title('my scatter')
```

```python
## Scatter plot with 3rd dimension
c = np.random.randint(0, 10, len(y))
plt.figure(figsize = (6, 4))
plt.scatter(y[:, 1], y [:, 0], c = c, cmap = 'coolwarm', marker = 'o')
plt.colorbar()
plt.title('my scatter')
```

```python
## Plotting histograms
plt.figure(figsize = (6, 4))
plt.hist(y, label = ['data1', 'data2'], bins = 15)
plt.legend(loc = 0)
plt.xlabel('value')
plt.ylabel('frequencies')
plt.title('my histo')
```

```python
## Plotting boxplots
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(1, 2, 1000)
y = np.column_stack((x1, x2)) ## combine since boxplot() takes 1 array as argument
fig, ax = plt.subplots(figsize = (6, 4))
box = plt.boxplot(y) ## plot 2 boxes
plt.setp(ax, xticklabels = ['data1', 'data2']) ## setp = set property
plt.xlabel('index')
plt.ylabel('data points')
plt.title('My boxes')

## changing the color of the box contours
for _, line_list in box.items():
    for line in line_list:
        line.set_color('b')
```

```python
## Maths inspired plot
## Defining the function
import matplotlib.pyplot as plt
import numpy as np
def fun(x):
    return .5 * np.exp(x) + 1
a, b = .5, 1.5 # integral limits
x = np.linspace(0, 2) # x-values for the plot
y = fun(x) # y-values for the plot
ix = np.linspace(a, b) # x-values within the integral limits
iy = fun(ix) # y-values within the integral limits
pts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)] # coordinates of points on the surface (polygon)

## The plot
from matplotlib.patches import  Polygon
fig, ax = plt.subplots(figsize = (10, 6)) # define figure and axis object
plt.plot(x, y, 'b', linewidth = 2) # plot the function
plt.ylim(bottom = 0) # min value for the y-axis
surf = Polygon(pts, facecolor = '.7', edgecolor = '.5') # plot the surface in grey
ax.add_patch(surf) # prev. cont.
plt.text(.5 * (a + b), 1, r'$\int_a^b f(x) \mathrm{d}x$',
        horizontalalignment = 'center', fontsize = 20) # adding the formula on the plot
plt.figtext(.9, .075, 'x') # axis labels
plt.figtext(.075, .9, 'f(x)')
ax.set_xticks((a, b)) # x-value labels
ax.set_xticklabels(('a', 'b'))
ax.set_yticks((fun(a), fun(b))) # y-value labels
ax.set_yticklabels(('f(a)', 'f(b)'))
```

## Static 3D Plotting


```python
## Simulated volatility surface
import numpy as np
strike = np.linspace(50, 150, 24) # ndarray with strike values
ttm = np.linspace(.5, 2.5, 24) # ndarray with maturities
strike, ttm = np.meshgrid(strike, ttm) # create a 2-dim ndarray (rectangular grid)
iv = (strike - 100) ** 2 / (100 * strike) / ttm # dummy implied volatility value

## The plot
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure(figsize = (10, 6))
ax = fig.gca(projection = '3d') # set up a canvas (container) for 3d plotting
surf = ax.plot_surface(strike, ttm, iv, rstride = 2, 
                       cstride = 2, cmap = plt.cm.coolwarm,
                      linewidth = .5, antialiased = True) # create the 3D plot
ax.set_xlabel('strike values') # set axis labels
ax.set_ylabel('time to maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink = .5, aspect = 5) # create a colorbar
```

```python
## Changing the view and using making a scatter
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(111, projection = '3d') # set up a canvas (container) for 3d plotting
ax.view_init(45, 45) # set the viewing angle
ax.scatter(strike, ttm, iv, zdir = 'z', s = 25, 
           c = 'b', marker = '^') # create the 3D scatter plot
ax.set_xlabel('strike values') # set axis labels
ax.set_ylabel('time to maturity')
ax.set_zlabel('implied volatility')
```

```python
# Data import
filename = '/content/EURUSD_D1.csv' # path of uploaded file
f = open(filename, 'r')
f.readlines()[:5] # Check data stracture 
```

```python
data = pd.read_csv(filename, index_col = 0, 
                   sep = "\t" , parse_dates = True)
data.info()
```

```python
# Visualize the data via multiple subplots
data.plot(figsize = (8, 10), subplots = True)
```

```python
# descriptive stats per column
data.describe().round(2)
```

```python
# Customization of the stats table output
data.aggregate([min, np.mean, np.std, np.median, max]).round(2)
```

```python
## Changes over time
```

```python
# absolute differences
data.diff().tail()
# aggregation operations could be applied in addition
```

```python
# percantage changes (simple returns)
data.pct_change().aggregate([np.mean, np.median]).round(3)
```

```python
# log returns
sdata = data.iloc[:, :4]
lret = np.log(sdata / sdata.shift(1))
lret.cumsum().apply(np.exp).plot(figsize = (8, 6)) 
```

```python
# Resampling
data.resample('1m', label = 'right').last().tail() 
# to avoid foresight bias make sure the right label of the 
# interval as well as the last data point in the interval 
# are used
```
